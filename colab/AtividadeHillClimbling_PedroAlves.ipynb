```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Dados Coletados ---

# Dados da 1ª simulação (Hill Climbing Tradicional) que você já tinha
resultados_tradicional = [924, 964, 997, 1019, 1021, 1042, 994, 1031, 970, 1037, 1013, 1037, 1024, 1037, 1037, 1007, 941, 986, 1037, 1037, 1042, 1024, 1031, 938, 964, 1037, 1017, 993, 1037, 1030]

# Dados da 2ª simulação (Stochastic Hill Climbing) que você acabou de gerar
resultados_stochastic = [967, 952, 822, 914, 901, 914, 853, 958, 871, 886, 940, 835, 935, 877, 769, 798, 960, 744, 832, 843, 751, 797, 859, 796, 971, 866, 862, 853, 878, 904]


# --- Análise e Gráfico 1: Apenas o Stochastic Hill Climbing ---

print("--- Análise dos Resultados do Stochastic Hill Climbing ---")
media_stochastic = np.mean(resultados_stochastic)
std_stochastic = np.std(resultados_stochastic)
print(f"Média do Fitness: {media_stochastic:.2f}")
print(f"Desvio Padrão do Fitness: {std_stochastic:.2f}\n")

# Configura o tema visual dos gráficos
sns.set_theme(style="whitegrid")

# Cria o primeiro gráfico
plt.figure(figsize=(8, 6))
sns.boxplot(y=resultados_stochastic, color="lightcoral")
plt.title('Resultados do Stochastic Hill Climbing (30 Execuções)', fontsize=16)
plt.ylabel('Qualidade (Fitness)', fontsize=12)
plt.show()


# --- Análise e Gráfico 2: Comparação entre os dois algoritmos ---

print("\n\n--- Análise Comparativa: Tradicional vs. Estocástico ---")
media_tradicional = np.mean(resultados_tradicional)
std_tradicional = np.std(resultados_tradicional)
print(f"Tradicional -> Média: {media_tradicional:.2f}, Desvio Padrão: {std_tradicional:.2f}")
print(f"Estocástico -> Média: {media_stochastic:.2f}, Desvio Padrão: {std_stochastic:.2f}")


# Preparar os dados para o gráfico comparativo usando a biblioteca Pandas
df_trad = pd.DataFrame({'Fitness': resultados_tradicional, 'Algoritmo': 'Tradicional'})
df_stoch = pd.DataFrame({'Fitness': resultados_stochastic, 'Algoritmo': 'Estocástico'})
dados_completos = pd.concat([df_trad, df_stoch])

# Cria o segundo gráfico (o comparativo)
plt.figure(figsize=(10, 8))
sns.boxplot(x='Algoritmo', y='Fitness', data=dados_completos, palette="pastel")
sns.stripplot(x='Algoritmo', y='Fitness', data=dados_completos, color=".25", size=5) # Adiciona os pontos individuais

plt.title('Comparação: Hill Climbing Tradicional vs. Estocástico', fontsize=16)
plt.xlabel('Tipo de Algoritmo', fontsize=12)
plt.ylabel('Qualidade da Solução (Fitness)', fontsize=12)
plt.show()
```
